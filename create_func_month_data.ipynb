# Import Libraries

```python
import pandas as pd
from datetime import datetime
from datetime import date, timedelta
data = pd.read_csv("data.csv")
data["month_name"] = (pd.to_datetime(data['date']).dt.month).apply(lambda x:datetime(2023, x, 1).strftime('%B'))
employee = 1000
```

# Total Calculation Function

```python
# Total
def total_day():
    totalDay = len(data["id"])
    return totalDay

def total_employee_worked():
    employee = sum(data["employee"])
    return employee

def total_mat_used():
    totalMat = sum(data["material"])
    return totalMat

def total_sr():
    totalSR = round(((sum(data["strike_rate"]))/(len(data["strike_rate"]))),2)*100
    return totalSR

def total_shop_covered():
    totalShop = sum(data["shop"])
    return totalShop

# Average
def total_avg_employee():
    totalAvgWork = round((sum(data["employee"]))/(len(data["employee"])))
    return totalAvgWork

def total_avg_mat():
    totalAvgMat = round((sum(data["material"]))/(len(data["material"])))
    return totalAvgMat

def total_avg_shop():
    totalAvgShop = round((sum(data["shop"]))/(len(data["shop"])))
    return totalAvgShop
```

# Maximum Data Query Function

```python
def max_employee():
    maxEmploee = max(data["employee"])
    maxEmployeeData = data.loc[data["employee"]==maxEmploee]
    d = str(maxEmployeeData["date"].to_string(index=False))
    day = d.replace("\n"," and ")
    maxData = f"From the whole database the maximum number of employee {maxEmploee} worked on {day}."
    return maxData

def max_mat():
    maxMat = max(data["material"])
    maxMatData = data.loc[data["material"]==maxMat]
    d = str(maxMatData["date"].to_string(index=False))
    day = d.replace("\n"," and ")
    maxData = f"maximum {maxMat} materials are found on date {day}"
    return maxData

def maxSR():
    maxSR = max(data["strike_rate"])
    maxSRData = data.loc[data["strike_rate"]==maxSR]
    d = str(maxSRData["date"].to_string(index=False))
    day = d.replace("\n"," and ")
    maxstrikerate = f"The maximum strike rate is {round((maxSR*100),2)} % found on date {day}."
    return maxstrikerate

def max_shop():
    maxShop = max(data["shop"])
    maxShopData = data.loc[data["shop"]==maxShop]
    d = str(maxShopData["date"].to_string(index=False))
    day = d.replace("\n"," and ")
    maxData = f"On {day} i found the maximum {maxShop} shops covered"
    return maxData
```
